# Community Detection with python-igraph

## This is a test interactive tutorial

The code below is completely AI generated


```python
import igraph as ig
import matplotlib.pyplot as plt

# Create a small example graph
g = ig.Graph.Famous("Zachary")  # Zachary's karate club network

# Apply community detection using the Louvain method
community = g.community_multilevel()

# Assign colors to nodes based on community membership
palette = ig.drawing.colors.ClusterColoringPalette(len(community))
colors = [palette[i] for i in community.membership]

# Plot the graph with community colors
fig, ax = plt.subplots(figsize=(8, 6))
ig.plot(
    g, 
    target=ax, 
    vertex_color=colors, 
    vertex_size=15, 
    edge_width=0.5, 
    layout=g.layout("fr")
)
plt.show()

# Print community structure
print("Detected Communities:")
for i, comm in enumerate(community, 1):
    print(f"Community {i}: {comm}")
```
